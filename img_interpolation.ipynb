```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.datasets import cifar10
```

```python
# load CIFAR-10 data
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

```python
sample_img = train_images[0]
```

```python
sample_img.shape
```

```python
# display the original image
plt.figure(figsize=(3, 3))
plt.imshow(sample_img)
plt.title('Original Image')
plt.show()
```

### Enlarging

```python
# save img as a file
image.save_img('sample_cifar_image.png', sample_img)

# resizing size (enlarging the image)
target_size = (200, 200)

# Interpolation methods
interpolation_methods = ['nearest', 'bilinear', 'bicubic', 'lanczos']
resized_images = []

# load and resize the image using different interpolation methods
for method in interpolation_methods:
    img = image.load_img('sample_cifar_image.png', target_size=target_size, interpolation=method)
    resized_img = image.img_to_array(img)
    print(f'({method}): {resized_img.shape=}')
    resized_images.append((method, resized_img))

lanczos_enlarged = img

# plotting the resized images in a 2x2 grid
plt.figure(figsize=(6, 6))

for i, (method, img) in enumerate(resized_images):
    plt.subplot(2, 2, i + 1)
    plt.imshow(img.astype('uint8'))
    plt.title(f'{method.capitalize()} Interpolation')

plt.tight_layout()
plt.show()

# display the original image
plt.figure(figsize=(3, 3))
plt.imshow(sample_img)
plt.title('Original Image')
plt.show()
```
